This notebook is used to display spectrogram from hdf5 dataset from each class

```python
import random
from ketos.data_handling.database_interface import open_file, open_table, load_audio
```

```python
db_save_path='../../saved_database/'
db_save_filename='db_test_3.0.h5'

# Open a connection to the database.
h5file = open_file(db_save_path+db_save_filename, 'r')
```

```python
def plot_n_spectrogram(group, number_of_spec):
    """ Plots n number of spectrograms randomly selected from a hdf5 group
     
        Args:
            group: str
                hdf5 database's group name.
            number_of_spec: int
                Number of spectrograms to plot from each group
    """
    
    table = open_table(h5file, group)
    specs_data = load_audio(table=table)
    print("Number of samples in ", group, ": ", len(specs_data))
    random_indices=random.sample(range(0,len(specs_data)), number_of_spec)
    for i in random_indices:
        specs_data[i].plot()
```

```python
plot_n_spectrogram('/CB50/Neg/data', 3)
plot_n_spectrogram('/CB50/BH/data', 3)
plot_n_spectrogram('/CB300/Neg/data', 3)
plot_n_spectrogram('/CB300/BH/data', 3)
```

```python
h5file.close()
```
